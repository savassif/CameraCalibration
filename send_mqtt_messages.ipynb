```python
dictionary_payload = {
    "object_id": 1,
    "x": 0.0,
    "y": 0.0,
    "header": {
        "stamp": {
            "sec": 10,
            "nsec": 890
        },
        "frame_id": "world"
    },
    "contour_points": [ 
        {
            "x": 0.1,
            "y": -0.1
        },
        {
            "x": 0.1,
            "y": 0.1
        },
        {
            "x": -0.1,
            "y": 0.1
        },
        {
            "x": -0.1,
            "y": -0.1
        }
    ],
    "bin_id": 1
}
```

```python
import json
import time

time_interval = 1

while True:
    dictionary_payload["object_id"] = time_interval
    json_file_as_string = json.dumps(dictionary_payload)
    client.loop()
    ret = client.publish(TOPIC, json_file_as_string, qos=1)
    print(f"Result: {ret.rc}")
    
    if (ret.rc == paho.MQTT_ERR_CONN_LOST or ret.rc == paho.MQTT_ERR_NO_CONN):
        ret = client.reconnect()
        ret = client.publish(TOPIC, json_file_as_string, qos=1)
        print(f"Result: {ret.rc}")
    
    time.sleep(time_interval) # Sleep for some seconds
    print(f"Waited for {time_interval} seconds")
    time_interval += 30
```
